```python
#Data Structure: [resturant_id, 2021, 2022, 2023, 2024]
import numpy as np
import matplotlib.pyplot as plt

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  #Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  #Beijing Bites
    [3, 200000, 230000, 260000, 300000],  #Pizza Hub
    [4, 180000, 210000, 240000, 270000],  #Burger Point
    [5, 160000, 185000, 205000, 230000]   #Chai Point
])

print("------Zomato Sales Analysis------")
print("\nSales Data Shape", sales_data.shape)
print("\nSample Data for First Three Resturants\n", sales_data[:3])

#total sales per year
print(np.sum(sales_data,axis=0))
```

```python
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("Total sales", yearly_total)
```

```python
#Minimum Sales Per Resturant
minimum_sales = np.min(sales_data[:, 1:], axis=1)
print("Minimum Sales:", minimum_sales)
```

```python
#Maximum Sales Per Year
max_sales = np.max(sales_data[:, 1:], axis=1)
print("Maximum Sales:", max_sales)
```

```python
#Average Sales Per Resturant
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average Sales per resturant", avg_sales)
```

```python
#Cumulative Sales
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("Cumulative Sales of the year\n", cumsum)
```

```python
#Plotting on Graph
plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative sales across all resturant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector Addition", vector1 + vector2)
print("\nMultiplication of Vector", vector1 * vector2)
print("\nDot Product", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("Angle: ", angle)
```
